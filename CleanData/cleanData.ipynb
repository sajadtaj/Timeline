## Set column names for EconomicCalender

```python
Data = EconomicCalender.copy()

Data.columns=['Date','Time','country','Volatility','Description','Evaluation of the event','--','Actual event','Evaluation data','Event forecast']
Data
```

## Select few Event 

```python
Data['Evaluation of the event'].unique()
```

```python
Data_usa         = Data[Data['country']=='United States                   ']
Data_usa_highVol = Data_usa[Data_usa['Volatility']=='High Volatility Expected        ']
Data_usa_highVol_Better = Data_usa_highVol[Data_usa_highVol['Evaluation of the event']=='Better Than Expected            ']
Data_usa_highVol_Better.drop_duplicates('Date',keep='first', inplace= True)
```

```python
Data_usa_highVol_Better
```

## Modify Data format to ready for merge

```python
Data_usa_highVol_Better['Date']= Data_usa_highVol_Better['Date'].apply(lambda x: x.replace("/","-"))
Data_usa_highVol_Better
```

## Remove space from desriotion

```python
def spaceRemove(x):
    return x.replace('  ','')
Data_usa_highVol_Better['Description'] = Data_usa_highVol_Better['Description'].apply(spaceRemove)
```

## Merge to Data frame 
>   Gold_Data   \
>   Data_usa_highVol_Better

```python
MergeData = pd.merge( Gold_Data,Data_usa_highVol_Better, on = ['Date'], how = 'left')
```

## Select spesific Date 

```python
temp1 = MergeData[MergeData['Date'] > "2019-01-01"	]
temp2 = temp1[temp1['Date'] < "2021-04-23"	]
temp2
```

```python
def convrtDate(x):
    return  datetime.strptime(x, r'%Y-%m-%d')

temp2['Date'].apply(convrtDate)
```
